```python
import jax
import jax.numpy as jnp
from jax import pmap

# 1. Create a function to run on each core
def multiply_by_two(x):
    return x * 2

# 2. Use pmap to distribute the function across all 8 TPU cores
parallel_fn = pmap(multiply_by_two)

# 3. Create data with 8 elements (one for each core)
data = jnp.arange(8)
result = parallel_fn(data)

print(f"Cores found: {jax.device_count()}")
print(f"Parallel Result: {result}")
```

```python
# 1. Distribute a simple function
parallel_add = pmap(lambda x: x + 100)

# 2. Create an array with exactly 8 elements (one for each TPU core)
data = jnp.arange(8)

# 3. Execute in parallel
result = parallel_add(data)

print("Input data: ", data)
print("Result from 8 cores: ", result)
# Each element in 'result' was processed by a different TPU core!
```
