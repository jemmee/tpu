```python
import time

# A complex function to give the compiler something to work with
@jax.jit
def complex_math(x):
    return jnp.sin(x) + jnp.cos(x) * jnp.exp(x)

x = jax.random.normal(jax.random.PRNGKey(0), (5000, 5000))

# FIRST RUN (Includes compilation time)
start = time.time()
_ = complex_math(x).block_until_ready()
print(f"First run (with JIT compile): {time.time() - start:.4f}s")

# SECOND RUN (Pure hardware execution)
start = time.time()
_ = complex_math(x).block_until_ready()
print(f"Second run (cached): {time.time() - start:.4f}s")
```
