```python
import numpy as np
import time

# 1. Create large random matrices (Standard 64-bit float)
size = 8192
a = np.random.randn(size, size).astype(np.float32)
b = np.random.randn(size, size).astype(np.float32)

print(f"Running MatMul on CPU ({size}x{size})...")

# 2. Benchmark the operation
# We use multiple iterations to get a stable average
iterations = 5
start_time = time.time()

for _ in range(iterations):
    # np.dot or @ performs the matrix multiplication
    result = np.dot(a, b)

end_time = time.time()

avg_time = (end_time - start_time) / iterations
print(f"Average time on CPU: {avg_time:.4f} seconds")
```
